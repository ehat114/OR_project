```python
fname = "Data/WoolworthsDemands.csv"

demandData = pd.read_csv(fname, index_col="Store").transpose()
demandData.index = pd.to_datetime(demandData.index)  
weekdays = demandData.loc[(demandData.index.weekday!=5) & (demandData.index.weekday!=6)]
saturdays = demandData.loc[demandData.index.weekday==5]
```

```python
weekdaysCD = np.concatenate([weekdays[store].tolist() for store in weekdays.loc[:, weekdays.columns.str.contains("Countdown") * ~weekdays.columns.str.contains("Metro")]])
weekdaysM = np.concatenate([weekdays[store].tolist() for store in weekdays.loc[:, weekdays.columns.str.contains("Metro")]])
weekdaysNCD = np.concatenate([weekdays[store].tolist() for store in weekdays.loc[:, ~weekdays.columns.str.contains("Countdown")]])
satCD = np.concatenate([saturdays[store].tolist() for store in saturdays.loc[:, saturdays.columns.str.contains("Countdown") * ~weekdays.columns.str.contains("Metro")]])
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

ax1.hist(weekdaysCD, density=True, histtype="stepfilled", bins=10)
ax1.set_title("Countdown Stores on Weekdays")

ax2.hist(weekdaysM, density=True, histtype="stepfilled", bins=10)
ax2.set_title("Countdown Metro Stores on Weekdays")

ax3.hist(weekdaysNCD, density=True, histtype="stepfilled", bins=10)
ax3.set_title("Non-Countdown Stores on Weekdays")

ax4.hist(satCD, density=True, histtype="stepfilled", bins=10)
ax4.set_title("Countdown Stores on Saturdays")

plt.show()
```

```python
demCD = demandData.loc[:, demandData.columns.str.contains("Countdown") * ~weekdays.columns.str.contains("Metro")].apply(np.mean, axis=1)
demM = demandData.loc[:, weekdays.columns.str.contains("Metro")].apply(np.mean, axis=1)
demS = demandData.loc[:, weekdays.columns.str.contains("SuperValue")].apply(np.mean, axis=1)
demF = demandData.loc[:, weekdays.columns.str.contains("FreshChoice")].apply(np.mean, axis=1)
```

```python
fig, ax1 = plt.subplots()

ax1.plot(demCD, label="Countdown")
ax1.plot(demM, label="Countdown Metro")
ax1.plot(demS, label="SuperValue")
ax1.plot(demF, label="FreshChoice")

ax1.set_ylabel("Avg. # of Pallets")
ax1.set_title("Demand Data")
ax1.legend()

plt.show()
```

```python
m1 = folium.Map(location=list(reversed(coords[2])), zoom_start=10)
m2 = folium.Map(location=list(reversed(coords[2])), zoom_start=5)
m3 = folium.Map(location=list(reversed(coords[2])), zoom_start=5)

for i, coord in enumerate(coords):
    if locations.Type[i] == "Countdown":
        iconCol = "green"
    elif locations.Type[i] == "FreshChoice":
        iconCol = "blue"
    elif locations.Type[i] == "SuperValue":
        iconCol = "red"
    elif locations.Type[i] == "Countdown Metro":
        iconCol = "orange"
    elif locations.Type[i] == "Distribution Centre":
        iconCol = "black"
    
    folium.Marker(list(reversed(coord)), popup=locations.Store[i], icon=folium.Icon(color = iconCol)).add_to(m1)
    folium.Marker(list(reversed(coord)), popup=locations.Store[i], icon=folium.Icon(color = iconCol)).add_to(m2)


    region = regions[locations.Store[i]]
    
    if region == 1:
        colour = "red"
    elif region == 2:
        colour = "blue"
    elif region == 3:
        colour = "green"
    elif region == 4:
        colour = "purple"
    else:
        colour = "orange"
    
    
    folium.Marker(list(reversed(coord)), popup=locations.Store[i], icon=folium.Icon(color=colour)).add_to(m3)
```
